```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
lm = np.genfromtxt("mean_little.txt", delimiter=" ")
ls = np.genfromtxt("std_little.txt", delimiter=" ") / np.sqrt(50)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(lm, c="red")
plt.fill_between(x=np.arange(len(lm)), y1=lm-ls, y2=lm+ls)
```
