```python
import matplotlib.pyplot as plt
import glob
import numpy as np

from IPython.display import display, clear_output
%matplotlib inline
```

```python
fig, ax = plt.subplots(1, figsize=(10, 5))

for m , s in zip(glob.glob('./mean_*.txt'), glob.glob('./std_*.txt')):
    lm = np.genfromtxt(m, delimiter=" ")
    ls = np.genfromtxt(s, delimiter=" ") / np.sqrt(50)
    
    label = 'Version: ' + m.split('\\')[1][5:].split('.txt')[0]
    line,  = ax.plot(lm/26, label=label)
    ax.fill_between(x=np.arange(len(lm)), y1=(lm-ls)/26, y2=(lm+ls)/26,
                    alpha=0.2, color=line.get_color())
    
    lm_list = list(lm)
    if len(lm_list) > 20:
        mean_lm = np.mean(sorted(lm_list)[-20:])/26
        ax.plot([0, 100], [mean_lm, mean_lm],'--', color=line.get_color(), alpha=0.3, linewidth=5)
        
    ax.set_xlim(0,100)
    ax.set_ylim(0,1)
ax.legend()
ax.set(**{'xlabel':'Epochs', 'ylabel':'Score'});
```

```python
import time
fig, ax = plt.subplots(1, figsize=(10, 5))


for _ in range(20):
    ax.cla()
    
    for m , s in zip(glob.glob('./mean_*.txt'), glob.glob('./std_*.txt')):
        
        lm = np.genfromtxt(m, delimiter=" ")
        ls = np.genfromtxt(s, delimiter=" ") / np.sqrt(50)
    
        label = 'Version: ' + m.split('\\')[1][5:].split('.txt')[0]
        line,  = ax.plot(lm/26, label=label)
        ax.fill_between(x=np.arange(len(lm)), y1=(lm-ls)/26, y2=(lm+ls)/26,
                        alpha=0.2, color=line.get_color())
    
        
        ax.set_xlim(0,100)
        ax.set_ylim(0,1)
    ax.legend()
    ax.set(**{'xlabel':'Epochs', 'ylabel':'Score'});
    
    display(fig)
    time.sleep(2)
    clear_output(wait=True)

```
