```python
import matplotlib.pyplot as plt
import glob
import numpy as np
%matplotlib inline
```

```python
fig, ax = plt.subplots(1, figsize=(10, 5))


for m , s in zip(glob.glob('./mean_*.txt'), glob.glob('./std_*.txt')):

    lm = np.genfromtxt(m, delimiter=" ")
    ls = np.genfromtxt(s, delimiter=" ") / np.sqrt(50)
    label = 'Version: ' + m.split('\\')[1][5:].split('.txt')[0]
    line,  = ax.plot(lm/26, label=label)
    ax.fill_between(x=np.arange(len(lm)), y1=(lm-ls)/26, y2=(lm+ls)/26,
                    alpha=0.2, color=line.get_color())
    ax.set_xlim(0,100)
    ax.set_ylim(0,1)
ax.legend()
ax.set(**{'xlabel':'Epochs', 'ylabel':'Score'});
```
